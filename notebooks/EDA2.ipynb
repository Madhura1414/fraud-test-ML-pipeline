```python
def get_age(age):
    if(age<=18):
        return "Child"
    elif(age>18 and age<=30):
        return "Youth"
    elif(age>30 and age<=40):
        return "Adult"
    elif(age>40 and age<=50):
        return "Mid Age"
    else:
        return "Senior"
    
df['age_group']=df['age'].apply(lambda x:get_age(x))
```
